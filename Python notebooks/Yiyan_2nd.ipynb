```python
income_zip_raw = pd.read_csv(datadir + '15zpallnoagi.csv')
pop_zip_raw = pd.read_csv(datadir + 'population.csv')
crime_raw = pd.read_csv(datadir + 'crime.csv')
```

```python
business_1000 = business.head(1000)
income_zip_1000 = income_zip_raw.head(1000)
pop_zip_1000 = pop_zip_raw.head(1000)
```

```python
new_col = []
for i in pop_zip_1000['Zip Code ZCTA']:
    new_col.append(str(i).zfill(5))
pop_zip_1000['Zip'] = pd.Series(new_col, index = range(1000))
```

```python
new_col_crime = []
ind_col_crime = []
for idx1 in range(len(crime_raw)):
    for idx2 in range(1000):
        if business_1000['state'][idx2] == crime_raw['states'][idx1]:
            new_col_crime.append(crime_raw['crime'][idx1])
            ind_col_crime.append(idx2)
```

```python
new_col_inc = []
ind_col_inc = []
new_col_pop = []
ind_col_pop = []
for idx1 in range(1000):
    for idx2 in range(1000):
        if business_1000['postal_code'][idx2] == str(income_zip_1000['ZIPCODE'][idx1]):
            new_col_inc.append(int(income_zip_1000['A00100'][idx1]))
            ind_col_inc.append(idx2)
        if business_1000['postal_code'][idx2] == pop_zip_1000['Zip'][idx1]:
            new_col_pop.append(str(pop_zip_1000['2010 Census Population'][idx1]))
            ind_col_pop.append(idx2)
```

```python
business_1000['income'] = pd.Series(new_col_inc, index = ind_col_inc)
business_1000['population'] = pd.Series(new_col_pop, index = ind_col_pop)
business_1000['crime'] = pd.Series(new_col_crime, index = ind_col_crime)
```
